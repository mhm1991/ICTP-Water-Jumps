**Step 1: Extraction of the hh and dipole vectors**

**Input:**  a trajectory file in .xyz format (in comments also for .pdb format).

**Output:** the *dipole* and *hh* vectors of every molecule along the simulation time. The name of the outputs is DP_space.dat and HH_space.dat


```python
from os import pathconf
import numpy as np

#Correct coordinates
def pbc(x,box_size):
  x = x - box_size*np.round(x/box_size)
  return x

#Examle for reading .pdb file
#!pip install biopython
#from Bio import PDB
#parser = PDB.PDBParser()
#io = PDB.PDBIO()
#struct = parser.get_structure('prd_space', '/content/drive/MyDrive/Angular_jump/prd_space.pdb')
#atom_symbol, coords = ([] for i in range (2))
#for model in struct:
#    for chain in model:
#        for residue in chain:
#            for atom in residue:
#                x,y,z = atom.get_coord()
#                coords.append([float(x),float(y),float(z)])
#coords_np = np.array(coords)

#Read .xyz file
file_path = '/content/drive/MyDrive/DeepMD/water_lmp_3057.xyz'
xyz_file = open(file_path)

N = xyz_file.readline()
header = xyz_file.readline()
atom_symbol, coords = ([] for i in range (2))

for line in xyz_file:
  try:
    atom, x, y, z = line.replace("  ", " ").split(" ")
    coords.append([float(x),float(y),float(z)])
  except:
    n = line.split("  ")
xyz_file.close()

dp_str = ""
hh_str = ""
dp_list = []
hh_list = []

coords_np = np.array(coords)

t_step = 1  #frequency (time delay between snapshots in your trajectory file)
num_molecul = 1019  #number of Oxygen molecules
box_size = 31.197

#Writing our dipole angles
for j in range(0,round(coords_np.shape[0]/(3*num_molecul)), t_step):
  for i in range(num_molecul):
    oh1 = coords_np[num_molecul*j*3 + i] - coords_np[num_molecul*(3*j+1) + 2*i]
    oh2 = coords_np[num_molecul*j*3 + i] - coords_np[num_molecul*(3*j+1) + 2*i + 1]
    oh1 = pbc(oh1, box_size)
    oh2 = pbc(oh2, box_size)

    dp = oh1 + oh2
    hh = oh1 - oh2

    dp_list.append(round(dp[0], 3))
    dp_list.append(round(dp[1], 3))
    dp_list.append(round(dp[2], 3))

    hh_list.append(round(hh[0], 3))
    hh_list.append(round(hh[1], 3))
    hh_list.append(round(hh[2], 3))

dp_np = np.array(dp_list)
hh_np = np.array(hh_list)


# open file in write mode
with open('/content/drive/MyDrive/DeepMD/DP_space_LAMMPS.dat', 'w') as fp:
    for item in dp_list:
        # write each item on a new line
        fp.write("%s  " % item)
    print('Done DP_space_LAMMPS.dat')

# open file in write mode
with open('/content/drive/MyDrive/DeepMD/HH_space_LAMMPS.dat', 'w') as fp:
    for item in hh_list:
        # write each item on a new line
        fp.write("%s  " % item)
    print('Done HH_space_LAMMPS.dat')
```

**Step 2: Angular jump detection protocol**

**input:** dp_space.dat or hh_space.dat

**output:** - gd_angle_vs_time_dp_1.dat -> it has 4 columns:
*  (i) the swing magnitude,
*  (ii) swing duration,
*  (iii) end time of swing,
*  (iv) number of molecule

**Parameters **

t_step = inverse of your sampling frequency (time delay between snapshots in your trajectory file)

num_mol = number of molecules

f_c = cutoff frequency (for low pass filter)

```python
# In this program I extract angular magnitudes and times for the molecules
import pandas as pd
import numpy as np
from scipy import signal
from scipy.signal import argrelextrema


def smooth(a, WSZ):
	# a: NumPy 1-D array containing the data to be smoothed
  # WSZ: smoothing window size needs, which must be odd number,
  # as in the original MATLAB implementation

	out0 = np.convolve(a,np.ones(WSZ,dtype=int),'valid')/WSZ
	r = np.arange(1,WSZ,2)
	start = np.cumsum(a[:WSZ-1])[::2]/r
	stop = (np.cumsum(a[:-WSZ:-1])[::2]/r)[::-1]
	return np.concatenate((start, out0, stop))

t_step_for_filt = 4 #frequency for filtering
t_step = 4 #frequency (time delay between snapshots in your trajectory file)
num_mol=1019
fc = 1/100
threshold = 1e-4

# here import the dipole/hh vector
A_from_file = pd.read_csv("/content/drive/MyDrive/DeepMD/DP_space_LAMMPS.dat", sep="  ", header=None)
A = np.zeros((3*num_mol, int(A_from_file.shape[1]/(3*num_mol))))
A1 = np.array(A_from_file)
for j in range (0, int(A_from_file.shape[1]/(3*num_mol))):
	A[:,j] = A1[0, j*3*num_mol : (j+1)*3*num_mol]

t_start = 500 #Starting time, because of equlibration
tim = A.shape[1] - t_start

b_time = []
b_ang  = []
b_dt   = []
b_mol  = []

counts = np.zeros([int(A.shape[0]/3)])

for j in range(0, num_mol):
  x = A[3*j+0, t_start : ]
  y = A[3*j+1, t_start : ]
  z = A[3*j+2, t_start : ]
  xyz2 = np.array([x[:],y[:],z[:]])
  xyz = np.transpose(xyz2)

  fs = 1/t_step_for_filt
  fc = 1/100
  # Apply butterworth filtering here to obtain the filtered vector (fx, fy, fz)
  [b,a] = signal.butter(2, fc/(fs/2))

  fxyz = np.zeros([len(x), 3])
  fxyz[:,0] = signal.filtfilt(b, a, x)
  fxyz[:,1] = signal.filtfilt(b, a, y)
  fxyz[:,2] = signal.filtfilt(b, a, z)

  sxyz = np.zeros([len(fxyz[:,0]), 3])
  sxyz[:,0] = smooth(fxyz[:,0], int(100./t_step))
  sxyz[:,1] = smooth(fxyz[:,1], int(100./t_step))
  sxyz[:,2] = smooth(fxyz[:,2], int(100./t_step))

  for ii in range (0, tim):
    sxyz[ii,:]=sxyz[ii,:]/np.sqrt(np.dot(sxyz[ii,:],sxyz[ii,:]))

  # Compute the velocity and dot product
  vxyz = np.zeros([len(fxyz[:,0])-1 , 3])
  vxyz[:,0] = np.diff(sxyz[:,0])
  vxyz[:,1] = np.diff(sxyz[:,1])
  vxyz[:,2] = np.diff(sxyz[:,2])

  sxyz_1 = np.zeros([len(fxyz[:,0])-1, 3])
  sxyz_1[:,0] = sxyz[:-1,0]
  sxyz_1[:,1] = sxyz[:-1,1]
  sxyz_1[:,2] = sxyz[:-1,2]

  qxyz = np.cross(sxyz_1, vxyz)

  for ii in range (0, len(qxyz)):
    qxyz[ii,:]=qxyz[ii,:]/np.sqrt(np.dot(qxyz[ii,:],qxyz[ii,:]))

	# Compute the dot product between consecutive frames as indication of change of geodesic.
  ah = np.zeros([len(qxyz)-1])

  for i in range (0, len(qxyz)-1):
    factor = np.dot(qxyz[i,:],qxyz[i,:])*np.dot(qxyz[i+1,:],qxyz[i+1,:])
    ah[i] = 1.0 - np.dot(qxyz[i,:],qxyz[i+1,:])/np.sqrt(factor)

    if(1e3*ah[i]>0.1):
      counts[j]+=1

    if(ah[i] <= threshold):
      ah[i] = 0.0

  l = argrelextrema(ah, np.greater)
  t = np.array(l)

  # change in jump of angle , change in the angle
  ang = np.zeros(int(len(l[0])))
  dt = np.zeros(int(len(l[0])))
  mol = np.zeros(int(len(l[0])))

  ang[0] = np.dot(xyz[t[0,0],:], xyz[0,:])/np.sqrt(np.dot(xyz[t[0,0],:],xyz[t[0,0],:])*np.dot(xyz[0,:],xyz[0,:]))
  ang[0] = np.arccos(ang[0])
  dt[0] = t[0,0]
  mol =  np.full(int(len(l[0])), j, dtype=int)

  for i in range(1, len(l[0])):
    ang[i] = np.dot(xyz[t[0,i],:],xyz[t[0,i-1],:])/np.sqrt( np.dot( xyz[t[0,i],:],xyz[t[0,i],:] ) * np.dot(xyz[t[0,i-1],:],xyz[t[0,i-1],:]) )
    ang[i] = np.arccos(ang[i])
    dt[i] = t[0,i]-t[0,i-1]

  ang = ang*(180/np.pi)

  b_ang.append(ang[:])
  b_dt.append(dt[:])
  b_time.append(t[0,:])
  b_mol.append(mol[:])

b_ang_array = []
b_dt_array = []
b_time_array = []
b_mol_array = []

for item in b_ang:
  for i in item:
    b_ang_array.append(i)

for item in b_dt:
  for i in item:
    b_dt_array.append(i * t_step)

for item in b_time:
  for i in item:
    b_time_array.append(i * t_step)

for item in b_mol:
  for i in item:
    b_mol_array.append(i)

tbs = np.array([b_ang_array[:], b_dt_array[:], b_time_array[:],b_mol_array[:]])
tbs_transpose = np.transpose(tbs)
tbs_sort = tbs_transpose[tbs_transpose[:,2].argsort()]

with open('/content/drive/MyDrive/DeepMD/tbs_sort_LAMMPS.txt', 'w') as f:
  np.savetxt(f, tbs_sort, fmt='%1.4e')
  print('Done tbs_sort.dat')
```


**Step 3: Local topology extraction**

**input:** the trajectory file in .xyz format

**output:** "defects_space.dat" contains the local topology of every molecule, i.e. the number of donors and acceptors.

parameters(internal variables to the code):
kg= number of molecules
numframes=number of frames
L=size of box

%Hydrogen bond definition
ang_ct= angular cutoff
dist_ct=Distance= distance cutoff

```python
import numpy as np
import math
from multiprocessing import Pool

def dis(vec1, vec2, L):
    vesc = (vec1 - vec2) - L * np.round((vec1 - vec2) / L)
    return np.sqrt(np.dot(vesc, vesc))

def angle(vec1, vec2, L):
    vec1 = vec1 - L * np.round(vec1 / L)
    vec2 = vec2 - L * np.round(vec2 / L)
    xa = np.dot(vec1, vec2) / np.sqrt(np.dot(vec1, vec1))
    xa = xa / np.sqrt(np.dot(vec2, vec2))
    if xa >= 1.0 or xa <= -1.0:
        xa = int(xa)
    res = math.acos(xa) * (180.0 / 3.1415926536)
    return res

def calculate_defects(args):
    coords_np, kg, ii, L, ang_ct, dist_ct = args
    defects = np.zeros([2 * kg])
    quickmatrix = np.zeros([kg + 1, kg + 1])

    for ig in range(1, kg):
        for jg in range(1 + ig, kg + 1):
            if dis(coords_np[(ig - 1) + ii * kg * 3, :], coords_np[(jg - 1) + ii * kg * 3, :], L) <= dist_ct:
                quickmatrix[ig, jg] = 1.0
                quickmatrix[jg, ig] = 1.0

    for ig in range(1, kg + 1):
        for jg in range(1, kg + 1):
            ang1 = angle(coords_np[(ig - 1) + ii * kg * 3, :] - coords_np[(jg - 1) + ii * kg * 3, :],
                         coords_np[(ig - 1) + ii * kg * 3, :] - coords_np[2 * (ig - 1) + ii * kg * 3 + kg, :], L)
            ang2 = angle(coords_np[(ig - 1) + ii * kg * 3, :] - coords_np[(jg - 1) + ii * kg * 3, :],
                         coords_np[(ig - 1) + ii * kg * 3, :] - coords_np[2 * (ig - 1) + 1 + ii * kg * 3 + kg, :], L)
            ang1 = min(ang1, ang2)

            if ang1 >= ang_ct:
                angw = 0
            else:
                angw = 1.0

            quickmatrix[ig, jg] = quickmatrix[ig, jg] * angw

    for i in range(1, kg + 1):
        defects[(i - 1) * 2] = sum(quickmatrix[i, :])
        defects[1 + (i - 1) * 2] = sum(quickmatrix[:, i])

    return defects

file_path = '/content/drive/MyDrive/DeepMD/water_lmp_3057.xyz'

with open(file_path, 'r') as xyz_file:
    N = xyz_file.readline()
    header = xyz_file.readline()
    atom_symbol, coords = ([] for _ in range(2))

    for line in xyz_file:
        try:
            atom, x, y, z = line.replace("  ", " ").split(" ")
            # atom_symbol.append(atom)
            coords.append([float(x), float(y), float(z)])
        except:
            n = line.split("  ")

coords_np = np.array(coords)

#number of Oxygen molecules
kg = 1019
#number of frames
numframes = int(len(coords) / (3 * kg))
#size of box
L = 31.197
#angular cutoff
ang_ct = 30
#distance cutoff
dist_ct = 3.5

pool = Pool()  # Create a process pool

args_list = [(coords_np, kg, ii, L, ang_ct, dist_ct) for ii in range(numframes)]
results = pool.map(calculate_defects, args_list)  # Parallel computation of defects

with open('/content/drive/MyDrive/DeepMD/defects_space_optimazed.txt', 'w') as f:
    for ii, defects in enumerate(results):
        np.savetxt(f, np.transpose(defects), fmt='%d', delimiter='', newline='           ')
        print(ii)
```

There is calculation of number of defects for Fig.3a [Correlation between the number of simultaneous large angular swings and the fluctuations in the local topology of the water H-bond network.](https://www.nature.com/articles/s41467-023-37069-9/figures/3)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

kg = 1019


defects = pd.read_csv("/content/drive/MyDrive/DeepMD/defects_space_optimazed.txt", sep="           ", header=None)
defects_array = np.array(defects)
defects_array = np.reshape(defects_array, (int(defects_array.shape[1]/(kg*2)),kg*2))

defects = np.ones([defects_array.shape[0],int(defects_array.shape[1]/2)])

for i in range(0,defects_array.shape[0]):
  for j in range(0,int(defects_array.shape[1]/2)):
    if (defects_array[i,2*j] == defects_array[i,2*j + 1] == 2):
       defects[i,j] = 0

total_defects_by_time = np.zeros([defects_array.shape[0]])
for i in range(0,defects_array.shape[0]):
  total_defects_by_time[i] = sum(defects[i,:])/(defects_array.shape[1]/2)

t_step = 1
amount = int(250 / t_step)

total_defects = np.zeros([int(defects_array.shape[0]/amount)])
for i in range(0,int(defects_array.shape[0]/amount)):
  total_defects[i] = sum(total_defects_by_time[amount*i:amount*(i+1)])/amount

with open('/content/drive/MyDrive/DeepMD/total_defects_space_optimazed.txt', 'w') as f:
  np.savetxt(f, total_defects, fmt='%1.4e')
  print('total_defects.dat')
```

**Step 4: Nearest hydrogen bonded neighbors**

**Input**: a trajectory in .xyz format, with frames 1ps apart from each other, and the output file of step 2 "gd_angle_vs_time_xx_1.dat"

**output**: the plots of the 1st 2nd 3rd 4th nearest neighbor distributions.

Ploting Fig. 5 [ Large angular swings occur close by in space.](https://www.nature.com/articles/s41467-023-37069-9/figures/5)
Fig 5.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import h5py
from sklearn.neighbors import NearestNeighbors
from scipy.stats import gaussian_kde

import scipy.io

# Custom distance function
def euclidean_with_periodicity(XI, XJ):
    periodic_factor = 31.197
    diff = (XI - XJ) - periodic_factor * np.round((XI - XJ) / periodic_factor)
    return np.sqrt(np.sum(diff ** 2))

# Load the data from file
dp = np.loadtxt("/content/drive/MyDrive/DeepMD/gd_angle_vs_time_dp_1_LAMMPS.dat")
hh = np.loadtxt("/content/drive/MyDrive/DeepMD/gd_angle_vs_time_hh_1_LAMMPS.dat")

dp_transpose = dp
hh_transpose = hh

# Load the xyz file
file_path = '/content/drive/MyDrive/DeepMD/water_lmp_3057.xyz'
xyz_file = open(file_path)
N = xyz_file.readline()
header = xyz_file.readline()
atom_symbol, coords = ([] for i in range (2))
for line in xyz_file:
  try:
    atom, x, y, z = line.replace("  ", " ").split(" ")
    coords.append([float(x),float(y),float(z)])
  except:
    n = line.split("  ")

coords_np = np.array(coords)

# Combine the data
yel = np.concatenate((dp_transpose[dp_transpose[:, 0] > 60], hh_transpose[hh_transpose[:, 0] > 60]), axis=0)
yel40 = np.concatenate((dp_transpose[dp_transpose[:, 0] > 40], hh_transpose[hh_transpose[:, 0] > 40]), axis=0)
yello = np.concatenate((dp_transpose[dp_transpose[:, 0] < 20], hh_transpose[hh_transpose[:, 0] < 20]), axis=0)

dp_transpose = np.transpose(dp)
hh_transpose = np.transpose(hh)

# First, second, third, and fourth nearest neighbors plot
w = 100
kl = 0

jumps_60 = [0,0,0,0]
rand_jumps_60 = [0,0,0,0]

for j in range(4):
    al = []
    al1 = []
    kl += 1
    for i in range(99):
        s = []
        coord = np.array(coords_np[i*1019*3*250 :(i*1019*3* 250 + 1019),:])
        #s = np.unique(yel[(yel[:, 2] < (i * 1000 + w)) & (yel[:, 2] > (i * 1000 - w)), 3])
        # Filter yel based on conditions
        filtered_yel = yel[(yel[:, 2] < i * 1000 + w) & (yel[:, 2] > i * 1000 - w)]
        s = np.unique(filtered_yel[:, 3])
        s = np.round(s).astype(int)

        # Generate random integers in the range [0, 1019]
        si = np.unique(np.random.randint(0, 1019, size=s.shape))

        # Perform k-nearest neighbors search if s is not empty
        if len(s) > 5:
            k = 5
            nbrs = NearestNeighbors(n_neighbors=k, metric=euclidean_with_periodicity)
            nbrs.fit(coord[si, :])  # Fit the model with the selected coordinates
            mIdx, mD = nbrs.kneighbors(coord[si, :])  # Find nearest neighbors for coord(si, :)
            al = np.concatenate((al, mIdx[:, kl ]))

            nbrs.fit(coord[s, :])  # Fit the model with the selected coordinates
            mIdx, mD = nbrs.kneighbors(coord[s, :])  # Find nearest neighbors for coord(s, :)
            al1 = np.concatenate((al1, mIdx[:, kl ]))

    jumps_60[j] = al1
    rand_jumps_60[j] = al


w = 100
kl = 0
jumps_40 = [0,0,0,0]
rand_jumps_40 = [0,0,0,0]

for j in range(4):
    al = []
    al1 = []
    kl += 1
    for i in range(99):
        s = []
        coord = np.array(coords_np[i*1019*3*250 :(i*1019*3*250 + 1019),:])
        # Filter yel based on conditions
        filtered_yel = yel40[(yel40[:, 2] < i * 1000 + w) & (yel40[:, 2] > i * 1000 - w)]
        s = np.unique(filtered_yel[:, 3])
        s = np.round(s).astype(int)

        # Generate random integers in the range [0, 1019]
        si = np.unique(np.random.randint(0, 1019, size=s.shape))

        # Perform k-nearest neighbors search if s is not empty
        if len(s) > 0:
            k = 5
            nbrs = NearestNeighbors(n_neighbors=k, metric=euclidean_with_periodicity)
            nbrs.fit(coord[si, :])  # Fit the model with the selected coordinates
            mIdx, mD = nbrs.kneighbors(coord[si, :])  # Find nearest neighbors for coord(si, :)
            al = np.concatenate((al, mIdx[:, kl ]))

            nbrs.fit(coord[s, :])  # Fit the model with the selected coordinates
            mIdx, mD = nbrs.kneighbors(coord[s, :])  # Find nearest neighbors for coord(s, :)
            al1 = np.concatenate((al1, mIdx[:, kl ]))


    jumps_40[j] = al1
    rand_jumps_40[j] = al


# Create a figure and subplot
fig, axs = plt.subplots(2, 2)

# Iterate through the subplots
for j in range(4):
    ax = axs[j // 2, j % 2]

    # Calculate the kernel density estimate (KDE) using scipy's gaussian_kde
    kde_al = gaussian_kde(jumps_60[j])
    xi = np.linspace(2, jumps_60[j].max(), 100)
    f_al = kde_al(xi)

    kde_al1 = gaussian_kde(rand_jumps_60[j])
    f_al1 = kde_al1(xi)

    # Plot the KDEs
    ax.plot(xi, f_al, 'r', linewidth=2, label='jumping molecules')
    ax.plot(xi, f_al1, 'k--', linewidth=2, label='random')

    # Set the legend, labels, and title for the subplot
    ax.legend(fontsize=6)
    ax.set_xlabel("Min. distance to jumping Molecule ($\AA$)", fontsize=8)
    ax.set_ylabel("PDF")
    ax.set_title("Nearest neighbor " + str(j + 1) + " for >60", fontsize=8)

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()


# Create a figure and subplot
fig, axs = plt.subplots(2, 2)

# Iterate through the subplots
for j in range(4):
    ax = axs[j // 2, j % 2]

    # Calculate the kernel density estimate (KDE) using scipy's gaussian_kde
    kde_al = gaussian_kde(jumps_40[j])
    xi = np.linspace(2, jumps_40[j].max(), 100)
    f_al = kde_al(xi)

    kde_al1 = gaussian_kde(rand_jumps_40[j])
    f_al1 = kde_al1(xi)

    # Plot the KDEs
    ax.plot(xi, f_al, 'r', linewidth=2, label='jumping molecules')
    ax.plot(xi, f_al1, 'k--', linewidth=2, label='random')

    # Set the legend, labels, and title for the subplot
    ax.legend(fontsize=6)
    ax.set_xlabel("Min. distance to jumping Molecule ($\AA$)", fontsize=8)
    ax.set_ylabel("PDF")
    ax.set_title("Nearest neighbor " + str(j + 1)  + " for >40", fontsize=8)

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```
