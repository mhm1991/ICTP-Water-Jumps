```python
import numpy as np
import matplotlib.pyplot as plt

corr300 = np.loadtxt('corr_croos_300.txt')
corr340 = np.loadtxt('corr_croos_340.txt')

plt.plot(corr300,label = '$T=300K$')
plt.plot(corr340,label = '$T=340K$')
plt.rcParams['figure.dpi'] = 400
plt.rcParams['savefig.dpi'] = 400
plt.rcParams['text.usetex'] = True

plt.xlabel('Lag Time ($\\tau$) [100fs]')
plt.ylabel('Correlation Function')
plt.title('DP-DP Correlation Function on N = 1019 Molecules')

#plt.title("Correlation Spectrum")
plt.grid(True)

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace this with your actual time series of correlation functions
# For demonstration purposes, let's generate a random time series
correlation_time_series300 = corr300#correlation[3800:0:-1]
correlation_time_series340 = corr340
total_lags = len(corr340)#correlation[3800:0:-1])

# Apply Fast Fourier Transform (FFT)
correlation_spectrum300 = np.fft.fft(correlation_time_series300)
correlation_spectrum340 = np.fft.fft(correlation_time_series340)


# Calculate the corresponding frequency values
sampling_rate = 1.0  # Adjust this if your time series has a different sampling rate
frequencies = np.fft.fftfreq(total_lags, d=1/sampling_rate)

size_freq = len(frequencies)//2

# Plot the spectrum
plt.plot(frequencies[:size_freq], np.abs(correlation_spectrum300)[:size_freq],label = "$T = 300K$")
plt.plot(frequencies[:size_freq], np.abs(correlation_spectrum340)[:size_freq],label = "$T = 340K$")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Correlation Spectrum")
plt.legend()
plt.grid(True)
plt.xlim([0,0.05])
plt.show()
```
